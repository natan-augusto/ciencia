This program is a mathematical application for quadratic equations; it functions as a root calculator for any quadratic equation with real roots.

---
# How to use:
- Simply insert the linear coefficient, angular coefficient, and the constant term, according to the general form of the quadratic equation: ax² + bx + c = 0.

---
NOTE: This application doesn't follow the official study guidelines.

```python
print("Importing module...")
import math
import time
print("Completed")
print()
def coefficient_a():
    a = float(input("Specify the quadratic coefficient (a): "))
    if a == 0:
        print()
        print("\033[31mInvalid structure for quadratic equation (a≠0)\033[0m")
        return coefficient_a()
    else:
        b = float(input("Specify the linear coefficient (b): "))
        c = float(input("Specify the independent term (c): "))
        quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x {str(c).replace('.', ',')}"
        if b>0 and c>0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(b).replace('.', ',')}x + {str(c).replace('.', ',')}"
        if b>0 and c<0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(b).replace('.', ',')}x {str(c).replace('.', ',')}"
        if b<0 and c>0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x + {str(c).replace('.', ',')}"
        if b<0 and c<0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x {str(c).replace('.', ',')}"
        if b>0 and c == 0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(b).replace('.', ',')}x"
        if b<0 and c == 0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(b).replace('.', ',')}x"
        if b == 0 and c == 0:
            quadratic_equation = f"{str(a).replace('.', ',')}x²"
        if b == 0 and c>0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² + {str(c).replace('.', ',')}"
        if b == 0 and c<0:
            quadratic_equation = f"{str(a).replace('.', ',')}x² {str(c).replace('.', ',')}"
        print()
        print(f"Quadratic equation to solve: \033[32m{quadratic_equation}\033[0m")
        print()
        print("Calculating the discriminant...")
        discriminant = (b**2) - 4 * a * c
        time.sleep(1)
        print(f"Discriminant calculated successfully: \033[32m{str(discriminant).replace('.', ',')}\033[0m")
        if discriminant < 0:
            time.sleep(1)
            print()
            print("\033[31mx doesn't exist in the set of real numbers.\033[0m")
            return coefficient_a()
        if discriminant == 0:
            time.sleep(1)
            print()
            print("It has coincident real roots.")
            print("\033[32mx exists in the set of real numbers.\033[0m")
            print()
            print("Calculating the roots...")
            time.sleep(1)
            root1 = (b * -1 + math.sqrt(discriminant))/(2*a)
            root2 = (b * -1 - math.sqrt(discriminant))/(2*a)
            print("\033[32mRoots calculated successfully.\033[0m")
            print()
            print(f"Root 1 (+): \033[32m{str(root1).replace('.', ',')}\033[0m")
            print(f"Root 2 (-): \033[32m{str(root2).replace('.', ',')}\033[0m")
            print("Enter the next equation:")
            return coefficient_a()
        if discriminant > 0:
            time.sleep(1)
            print()
            print("\033[32mx exists in the set of real numbers.\033[0m")
            print()
            print("Calculating the roots...")
            time.sleep(1)
            root1 = (b * -1 + math.sqrt(discriminant))/(2*a)
            root2 = (b * -1 - math.sqrt(discriminant))/(2*a)
            print("\033[32mRoots calculated successfully.\033[0m")
            print()
            print(f"Root 1 (+): \033[32m{str(root1).replace('.', ',')}\033[0m")
            print(f"Root 2 (-): \033[32m{str(root2).replace('.', ',')}\033[0m")
            print("Enter the next equation:")
            return coefficient_a()
coefficient_a()
```
